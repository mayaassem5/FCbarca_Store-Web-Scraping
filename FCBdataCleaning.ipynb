```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\original\Downloads\FCBshop.csv')
df.head()
```

```python
df['price'] = df['price'].str.replace(',', '')
df['price'] = df['price'].str.replace('LE ', '')
df['price'] = df['price'].str.replace('EGP', '')

df.price.head(20)
```

```python
df['price'] = df['price'].astype(float)
df.info()
```

```python
df_cleaned = df.drop_duplicates()
df_cleaned.shape
```

```python
df_cleaned['product_collection'] = df_cleaned.product_collection.str.capitalize()
df_cleaned
```

```python
stadium_collection = df_cleaned.loc[df_cleaned['product_collection'] == 'Stadium collection', :].sort_values('price', ascending=False)
stadium_collection.head()
```
